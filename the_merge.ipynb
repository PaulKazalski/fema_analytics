```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('DisasterDeclarationsSummaries2.csv')
df2 = pd.read_csv('FEMA_Housing_Assistance_Owners2.csv')
df3 = pd.read_csv('FEMA_Housing_Assistance_Renters2.csv')
```

```python
df1 = df1.rename(columns ={"disasterNumber" : "Disaster"})
df1.columns
```

```python
df1 = df1.loc[:,["address", "Disaster", "fyDeclared", "incidentType"]]
df1 = df1.loc[(df1 ["Disaster"]>=4116), :]
df1 = df1.loc[(df1 ["Disaster"]<=4466), :] 
df1 = df1.sort_values("Disaster")
df1
```

```python
df2 = df2.rename(columns ={"Total Approved IHP Amount" : "IHP Amount for Renters"})
df2 = df2.rename(columns ={"Approved for FEMA Assistance" : "Renters Approved"})
df2 = df2.rename(columns ={"Approved between $1 and $10,000" : "Renters <$10k"})
df2 = df2.rename(columns ={"Approved between $10,001 and $25,000" : "Renters $10-25k"})
df2 = df2.rename(columns ={"Approved between $25,001 and Max" : "Renters $25k - Max"})
df2 = df2.rename(columns ={"Total Max Grants" : "Renters Max"})
df2                 
```

```python
df2.columns
```

```python
df2a = df2.loc[:,["address", "Disaster", "IHP Amount for Renters","Renters Approved", 
                 "Renters <$10k", "Renters $10-25k", "Renters $25k - Max", "Renters Max"]]
del(df2a["Disaster"])
df2a['IHP Amount for Renters'] = df2a['IHP Amount for Renters'].str.replace(',','').str.replace('$','').astype('int')
df2a['Renters Approved'] = df2a['Renters Approved'].str.replace(',','').str.replace('$','').astype('int')
df2a['Renters <$10k'] = df2a['Renters <$10k'].str.replace(',','').str.replace('$','').astype('int')
df2a['Renters $10-25k'] = df2a['Renters $10-25k'].str.replace(',','').str.replace('$','').astype('int')
df2a['Renters $25k - Max'] = df2a['Renters $25k - Max'].str.replace(',','').str.replace('$','').astype('int')
df2a['Renters Max'] = df2a['Renters Max'].str.replace(',','').str.replace('$','').astype('int')

df2a.dtypes
```

```python
df2b = df2a.groupby(["address"])

df2b = df2b.sum()
df2b
```

```python
#result = pd.merge(left, right, how='left', on=['key1', 'key2'])
merge23 = pd.merge(df1, df2b, how = 'outer', on=['address'])
merge23
```

```python
dfs = [df1,df2b]
all_data_df = pd.concat(dfs, axis = 1,join = 'outer',keys = 'address', ignore_index = True)
all_data_df
```

```python
df3 = df3.rename(columns ={"Total Approved IHP Amount" : "IHP Amount for Owners"})
df3 = df3.rename(columns ={"Approved for FEMA Assistance" : "Owners Approved"})
df3 = df3.rename(columns ={"Approved between $1 and $10,000" : "Owners <$10k"})
df3 = df3.rename(columns ={"Approved between $10,001 and $25,000" : "Owners $10-25k"})
df3 = df3.rename(columns ={"Approved between $25,001 and Max" : "Owners $25k - Max"})
df3 = df3.rename(columns ={"Total Max Grants" : "Owners Max"})
```

```python
df3a = df3.loc[:,["address", "Disaster", "IHP Amount for Owners","Owners Approved", 
                 "Owners <$10k", "Owners $10-25k", "Owners $25k - Max", "Owners Max"]]
del(df3a["Disaster"])

df3a.dtypes
```

```python
df3a
```

```python
df3b = df3a.groupby(["address"])

df3b = df3b.sum()
df3b
```

```python
df3.columns
```

```python
dfs = [df1,df2,df3]
all_data_df = pd.concat(dfs, axis = 1,join = 'outer',keys = 'address', ignore_index = True)
all_data_df.head()
```
